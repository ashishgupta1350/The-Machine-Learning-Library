```python
from sklearn import datasets as ds
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression as LOR
from sklearn import model_selection as ms
```

```python
cancer=ds.load_breast_cancer()
```

```python
X=cancer.data
Y=cancer.target
df=pd.DataFrame(X)
df.columns=cancer.feature_names
df.describe()
```

```python
clf=LOR() # clf is the classifier
clf.fit(X,Y)

```

```python
clf.score(X,Y)
```

```python
Checking_data=clf.predict_proba(X)
```
